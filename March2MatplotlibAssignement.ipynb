## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt 
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

### scatter plot creates graph which shows only points in the graph for each (x,y) pair. This function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

```python
plt.scatter(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
## Use the following data:


    import numpy as np
    For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
    For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
    For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
    For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

### With the subplot() function you can draw multiple plots in one figure
### The subplot() function takes three arguments that describes the layout of the figure.
### The layout is organized in rows and columns, which are represented by the first and second argument.
### The third argument represents the index of the current plot.

```python
import numpy as np
import matplotlib.pyplot as plt
## line 1
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("line 1")
## line 2
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title("line 2")

## line 3
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("line 3")
## line 4
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title("line 4")

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

### Bar plot represent data visualization in the form of bars. Each bar represents Data at Y axis and category at x axis.
### The categories and their values represented by the first and second argument as arrays.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("company")
plt.ylabel("Profit")
plt.show()
# horizontal bar
plt.barh(company,profit)
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
    box1 = np.random.normal(100, 10, 200)
    box2 = np.random.normal(90, 20, 200)

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

### A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum, first quartile, median, third quartile and maximum. In the box plot, a box is created from the first quartile to the third quartile, a vertical line is also there which goes through the box at the median. Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.

```python
data = [box1, box2]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
